```python
'''
Read cGENIE .res files via names
'''
import os
import numpy as np
import pandas as pd
directory = '/Volumes/DA/cGENIE/ML.petm/petm020.EXAMPLE.worbe2.RidgwellHargreaves1997_S36x36/'
foldi = '/biogem/'
file1 = 'biogem_series_atm_pCO2.res'
file2 = 'biogem_series_ocn_ALK.res'
file3 = 'biogem_series_ocn_DIC.res'
file4 = 'biogem_series_sed_CaCO3.res'
savename = 'pCO2vsALK.xlsx'
# all subfolders
# [x[0] for x in os.walk(directory)]
directory_contents = os.listdir(directory)
dirn = int(len(directory_contents)/2 -1)
result = np.full((dirn,5),np.nan)
i = 0

for item in directory_contents:
    if item[-4:] not in ['r.gz','.xls','xlsx']:
        print(item)
        outgas = int(item[-8]) + int(item[-6])/10
        result[i,0] = outgas
        
        #print(outgas)
        filename1 = directory + item + foldi + file1
        filename2 = directory + item + foldi + file2
        filename3 = directory + item + foldi + file3
        filename4 = directory + item + foldi + file4
        
        with open(filename1) as f:
            for line in f:
                pass
            lastline = line
            print(lastline)
        result[i,1] = lastline[-13:-1]
        
        with open(filename2) as f:
            for line in f:
                pass
            lastline = line
            print(lastline)
        result[i,2] = lastline[34:46]
        
        with open(filename3) as f:
            for line in f:
                pass
            lastline = line
            print(lastline)
        result[i,3] = lastline[34:46]
        
        with open(filename4) as f:
            for line in f:
                pass
            lastline = line
            print(lastline)
        result[i,4] = lastline[-10:-1]
        
        i += 1
        
result.sort(axis = 0)
result[:,1] = result[:,1] * 1E+6
print(result)
fullname = directory + savename
column_values = ['outgassing rate (x 5.58E+12 mol C yr-1)','pCO2 (ppm)','global mean ALK (mol kg-1)','global mean DIC (mol kg-1)','mean CaCO3 composition (wt%)']
df = pd.DataFrame(data = result,
                  columns = column_values) 

df.to_excel(fullname)
```
